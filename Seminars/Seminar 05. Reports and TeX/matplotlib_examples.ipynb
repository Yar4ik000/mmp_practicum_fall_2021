### Bar plot (.bar)

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

plt.title("Сравнение эффективности моделей по времени", fontsize=20)
plt.xlabel("Распределения", fontsize=18)
plt.ylabel("Время, ms", fontsize=18)

data = np.random.uniform(0, 100, (3,5))
names = ["hi", "i'm", "fine", "Thank you", "(ohhh)"]

w = 0.2
pos = np.arange(len(names))
ax.set_xticks(pos)
ax.set_xticklabels(names, rotation=45, fontsize=18)
ax.grid()

ax.bar(pos-w, data[0], width=w, color="crimson", label="Модель 1", log=False)
ax.bar(pos,   data[1], width=w, color="darkcyan",  label="Модель 2", log=False)
ax.bar(pos+w, data[2], width=w, color="darkorange",  label="Модель 3", log=False)
plt.ylim(-3,150)

ax.legend(loc=1, fontsize=18)

plt.show()
fig.savefig("times.pdf", bbox_inches = 'tight')
```

```python
with plt.style.context("seaborn-dark"):

    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(111)
    
    plt.title("Сравнение эффективности моделей по времени", fontsize=20)
    plt.xlabel("Распределения", fontsize=18)
    plt.ylabel("Время, ms", fontsize=18)

    data = 10**np.random.uniform(0, 5, (3,5))
    names = ["hi", "i'm", "fine", "Thank you", "(ohhh)"]

    w = 0.2
    pos = np.arange(len(names))
    ax.set_xticks(pos)
    ax.set_xticklabels(names, rotation=45, fontsize=18)

    ax.bar(pos-w, data[0], width=w, color="crimson", label="Модель 1", log=True)
    ax.bar(pos,   data[1], width=w, color="darkcyan",  label="Модель 2", log=True)
    ax.bar(pos+w, data[2], width=w, color="darkorange",  label="Модель 3", log=True)
    plt.ylim(1,(100*data.max()))
    ax.legend(loc=1, fontsize=18)
    plt.grid()

    plt.show()
    fig.savefig("times.pdf", bbox_inches = 'tight')
```

### Images (.imshow)

```python
mx1 = np.zeros((1000,1000))
mx2 = np.zeros((1000,1000))
for i1 in range(mx1.shape[0]):
    for i2 in range(mx1.shape[1]):
        x, y = (i1 - 500) / 100, (i2 - 500) / 100
        mx1[i1, i2] = np.sin(x) + np.cos((x - y)**2)
        mx2[i1, i2] = (y**2 - x**3) * (1 + np.sin(x**2 + y**2))
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plt.title("Something 1", fontsize="18")
plt.axis('off')
plt.imshow(mx1, cmap="inferno")
plt.plot()

ax = fig.add_subplot(122)
plt.title("Something 2", fontsize="18")
plt.axis('off')
plt.imshow(mx2, cmap="RdPu")
plt.plot()

fig.savefig("inferno.pdf", bbox_inches = 'tight')
```

### Default plots (.plot)

```python
data = np.arange(3000, 0, -1).reshape(3,1000)
data[1] -= 1100
data[0] -= 1800
data = data / 50
data += np.sin(data)
data = 1.1 ** data
```

```python
with plt.style.context("seaborn-dark"):
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(111)
    
    plt.title("Сравнение эффективности моделей по времени", fontsize=20)
    plt.xlabel("Распределения", fontsize=18)
    plt.ylabel("Время, ms", fontsize=18)
    
    ax.set_xticklabels(pos, rotation=45, fontsize=14)
    ax.set_yticklabels(np.arange(10), rotation=45, fontsize=14)
    pos = np.arange(data.shape[1]) * 10
    lw = 2
    plt.plot(pos, data[0], color="darkblue", linewidth=lw, label="Model 1")
    plt.plot(pos, data[1], color="crimson", linewidth=lw, label="Model 2")
    plt.plot(pos, data[2], color="green", linewidth=lw, label="Model 3")
    ax.legend(loc=0, fontsize=18)
    plt.grid()
    #plt.ylim(-10,400)
    
    plt.show()
    fig.savefig("times.pdf", bbox_inches = 'tight')
```

### Set of dots (.scatter)

```python
def f(x,y):
    return 1 - (x**2 + (y - ((x**2)**(1/3)))**2)
```

```python
mx = np.zeros((100,100))
for i1 in range(mx.shape[0]):
    for i2 in range(mx.shape[1]):
        x, y = (i1 - 50) / 30 * 1.15, (i2 - 50) / 30 * 1.5
        mx[i1,i2] = f(x, y)
```

```python
x1,y1 = np.where(mx < 0)
x2,y2 = np.where(abs(mx) < 1)
x3,y3 = np.where(mx > 0)
```

```python
fig = plt.figure(figsize=(8, 8))
plt.title("Points", fontsize=18)
plt.xlabel("x-coordinates", fontsize=16)
plt.ylabel("y-coordinates", fontsize=16)

scale = 3
plt.scatter(x1,y1, color="aqua", s=scale)
plt.scatter(x2,y2, color="darkviolet", s=scale)
plt.scatter(x3,y3, color="orchid", s=scale)
```
